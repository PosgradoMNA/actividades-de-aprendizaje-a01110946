### Paso 6: Escala los datos, si consideras necesario

```python
# Antes de escalar los datos, primero seleccionamos solamente las variables numéricas
df_numerical = df[['X1', 'X12', 'X13', 'X14', 'X15', 'X16', 'X17', 'X18', 'X19', 'X20', 'X21', 'X22', 'X23']]
print(df_numerical)
```

```python
from sklearn.preprocessing import MinMaxScaler
MinMaxEsc = MinMaxScaler(feature_range=(0,1)).fit(df_numerical)

df_numerical_scaled = pd.DataFrame(MinMaxEsc.transform(df_numerical))
print(df_numerical_scaled)
```

### Paso 7: Reduce las dimensiones con PCA, si consideras necesario.

```python
from sklearn.decomposition import PCA
pca=PCA()
pca_v = pca.fit_transform(df_numerical_scaled)

# Mostramos los primeros 10 valores
pca_v[:10]
```

Indica la varianza de los datos explicada por cada componente seleccionado. Para actividades de exploración de los datos la varianza > 70%

```python
# Obtenenemos la varianza explicada por componente y la varianza acumulada
pca_df_numerical_scaled = pd.DataFrame({
'Varianza explicada%': np.round(pca.explained_variance_ratio_,4) * 100,
'Varianza acumulada%': np.cumsum(pca.explained_variance_ratio_) * 100})
pca_df_numerical_scaled
```

### Paso 8: Elabora los histogramas de los atributos para visualizar su distribución

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.set(rc={'figure.figsize':(17,12)})
fig, axes = plt.subplots(6, 5)    
  
for k in range(len(df_numerical_scaled)):       
     
    plt.subplot(13,2,k+1) 
    plt.hist(df_numerical_scaled, bins=20)                                       
    plt.xlabel(df_numerical_scaled[k])
    if k==0:
      plt.ylabel('Gráficas')
```
